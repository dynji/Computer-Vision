```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def corr2d(X, K):
    h, w = K.shape
    Y = np.zeros((X.shape[0] - h + 1, X.shape[1] - w + 1))
    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            Y[i, j] = (X[i: i + h, j: j + w] * K).sum()
    return Y
```

```python
def padding(X):
    #h= X.shape[0]-3+1
    z = np.zeros((224,1), dtype = X.dtype)

    P = np.concatenate((X, z), axis = 1)
    P = np.concatenate((z, P), axis =1)

    z = np.zeros((1,226), dtype = P.dtype)
    P = np.concatenate((z, P))
    P = np.concatenate((P, z))
    X = P
    return X
```

```python
def Conv2D(path,K):
  image = cv.imread(path)

  def corr2d(X, K):
    h, w = K.shape
    Y = np.zeros((X.shape[0] - h + 1, X.shape[1] - w + 1))
    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            Y[i, j] = (X[i: i + h, j: j + w] * K).sum()
    return Y

  def padding(X):
    #h= X.shape[0]-3+1
    z = np.zeros((224,1), dtype = X.dtype)
    P = np.concatenate((X, z), axis = 1)
    P = np.concatenate((z, P), axis =1)
    z = np.zeros((1,226), dtype = P.dtype)
    P = np.concatenate((z, P))
    P = np.concatenate((P, z))
    X = P
    return X
  image_arr = np.array(image)
  B = padding(image_arr[:,:,0])
  B = corr2d(B, K)
  G = padding(image_arr[:,:,1])
  G = corr2d(G, K)
  R = padding(image_arr[:,:,2])
  R = corr2d(R, K)

  Z = np.stack([B, G, R], axis = 2)
  Z = Z.astype(np.uint8)

  plt.imshow(Z)
  plt.show
```
